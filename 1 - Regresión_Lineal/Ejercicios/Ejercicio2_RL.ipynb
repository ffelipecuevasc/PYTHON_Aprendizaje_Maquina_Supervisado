```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn import linear_model as lm

print(F"隆Importaci贸n correcta de Pandas! Versi贸n: {pd.__version__}")
print(F"隆Importaci贸n correcta de Numpy! Versi贸n: {np.__version__}")
print(F"隆Importaci贸n correcta de Matplotlib! Versi贸n: {plt.matplotlib.__version__}")
print(F"隆Importaci贸n correcta de SK Learn! Versi贸n: {sklearn.__version__}")
```

```python
# Dataset de 20 registros simulados
datos_consumo = {
    "Motor":     [1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1,
                  1.2, 1.4, 1.6, 1.8, 2.0, 2.2, 1.5, 1.7, 1.9, 2.1],
    "Velocidad": [100] * 20,  # Constante
    "Precio":    [1300, 1320, 1340, 1360, 1380, 1400, 1420, 1440, 1460, 1480,
                  1350, 1370, 1390, 1410, 1430, 1450, 1470, 1490, 1500, 1520],
    "Consumo":   [18, 17, 16.5, 16, 15, 14.5, 14, 13.5, 13, 12.5,
                  17.5, 16.8, 15.5, 14.2, 13.8, 12.8, 15.8, 14.8, 13.2, 12.2]
}

df = pd.DataFrame(datos_consumo)
df
```

```python
X = df[["Motor", "Velocidad", "Precio"]]  # Variables independientes
y = df["Consumo"]                         # Variable dependiente

modelo = lm.LinearRegression()
modelo.fit(X, y)
```

```python
# Solicitar datos al usuario
distancia = float(input("Ingrese la distancia del viaje (km): "))
motor = float(input("Ingrese el tama帽o del motor (ej: 1.6): "))
precio_combustible = float(input("Ingrese el precio del combustible por litro (CLP): "))

# La velocidad es constante
velocidad = 100

# Impresi贸n de datos ingresados
print(F"Los datos ingresados fueron: \nDistancia = {distancia}"
      F"\nMotor = {motor}"
      F"\nPrecio = {precio_combustible}"
      F"\nVelocidad = {velocidad}")
```

```python
# predict() siempre devuelve una lista de valores predichos, incluso si hay solo uno
# Por eso, al final se especifica [0] para que devuelva el primer y 煤nico elemento en formato FLOAT
# De lo contrario, devolver铆a una lista (serie) de valores, algo complejo para c谩lculos posteriores
consumo_predicho = modelo.predict(pd.DataFrame({"Motor": [motor],
                                                "Velocidad": [velocidad],
                                                "Precio": [precio_combustible]}))[0]
print(f"\nConsumo estimado del veh铆culo: {consumo_predicho:.2f} km/l")
```

```python
gasto_total = (distancia / consumo_predicho) * precio_combustible
print(f"Gasto estimado de combustible para el viaje: ${gasto_total:,.0f} CLP")
```
